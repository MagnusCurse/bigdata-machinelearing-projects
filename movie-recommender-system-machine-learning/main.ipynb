```python
import pandas as pd
```

```python
movies = pd.read_csv('dataset.csv')
```

```python
movies = movies[['id', 'title', 'overview', 'genre']]
movies
```

```python
movies['tags'] = movies['overview'] + movies['genre']
movies
```

```python
movies = movies.drop(columns=['overview', 'genre'])
movies
```

```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
# max_features=10000: this parameter limits the number of features (or words) to the 10,000 most frequent ones across all documents in your dataset
# stop_words='english': this parameter tells the CountVectorizer to ignore common English stop words such as "the", "and", "is", etc.
cv = CountVectorizer(max_features=10000, stop_words='english')
```

```python
# Convert the 'tags' column to a NumPy array of strings, where each elements represent a 'tags' text.
tags_array = movies['tags'].values.astype('U')

# Transform the text data into a document-term-matrix (which will have a lot of zero value)
# Then transform this document-term-matrix into a sparse matrix and return it.
vector_sparse = cv.fit_transform(tags_array)

# Convert the sparse matrix to a dense NumPy array.
vector = vector_sparse.toarray()
```

```python
vector
```

```python
# Get the shape (dimensions) of the 'vector' array
# This returns a tuple where the first element is the number of rows (documents)
# and the second element is the number of columns (features/words in the vocabulary)

vector.shape
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity = cosine_similarity(vector)
```

```python
similarity  # ? why having so many cosine similarity, what every cosine similarity represent ?
```

```python
movies[movies['title'] == "The Godfather"].index[0]  # Get the index of one of the movies by title
```

```python
# Print all the movies titles order by cosine similarity
# key=lambda vector:vector[1]: sort according to the second column of list(enumerate(similarity[2])), this 'vector' is not the vector above, just key word
distances = sorted(list(enumerate(similarity[2])), reverse=True, key=lambda vector:vector[1])

# movies.iloc[distance[0]]: The iloc method is a way to index the movies DataFrame by integer position.
for distance in distances:
    print(movies.iloc[distance[0]].title)
```
